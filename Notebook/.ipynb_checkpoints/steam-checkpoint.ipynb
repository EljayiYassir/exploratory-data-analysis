<h4 id="heading"><span style="background-color:#fcf3b9; color:#3d3710 ; display:fill;border-radius:5px; font-family:Georgia">🗄️ Outliers values </span></h4>


```python
# Create box plots for numerical columns
numerical_columns = ["Income", "Children", "Cars", "Age"]
raw_df[numerical_columns].boxplot()
plt.title("Box Plot for Numerical Columns")
plt.show()
```

```python
# Assuming you have a DataFrame named df with the four numerical columns: "Income", "Children", "Cars", and "Age"
numerical_columns = ["Income", "Children", "Cars", "Age"]

# Create four separate side-by-side box plots
fig, axes = plt.subplots(1, 4, figsize=(20, 6))
for i, column in enumerate(numerical_columns):
    raw_df.boxplot(column=column, ax=axes[i], patch_artist=True)
    axes[i].set_title(f"Box Plot for {column}")
    axes[i].set_ylabel("Value")

plt.show()
```

```python
# Calculate Z-scores for numerical columns
z_scores = stats.zscore(raw_df[numerical_columns])

# Find outliers based on a Z-score threshold (e.g., 3)
outlier_threshold = 3
outliers = (abs(z_scores) > outlier_threshold).any(axis=1)

# Print the rows with outliers
print('The number of outliers with Z-scores in this dataset is:', len(raw_df[outliers]))
```

```python
# Calculate the IQR for numerical columns
Q1 = raw_df[numerical_columns].quantile(0.25)
Q3 = raw_df[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Find outliers based on the IQR method
outliers = ((raw_df[numerical_columns] < (Q1 - 1.5 * IQR)) | (raw_df[numerical_columns] > (Q3 + 1.5 * IQR))).any(axis=1)

# Print the rows with outliers
print('The number of outliers with IQR method in this dataset is:' ,raw_df[outliers].shape[0])
# raw_df[outliers]
```

<h4 id="heading"><span style="background-color:#fcf3b9; color:#3d3710 ; display:fill;border-radius:5px; font-family:Georgia">📊 Bar plot using matplotlib </span></h4>


```python
# print a bar plot for all the categorical columns

# Create a costume color for each columns
custom_colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray']

# Create subplots with the desired layout
num_plots = len(categorical_columns)
num_rows = num_plots // 2 + num_plots % 2
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 4 * num_rows))

# Flatten the axes array in case we have only one row of subplots
axes = axes.flatten()

# Loop through categorical columns and create bar plots
for i, column in enumerate(categorical_columns):
    raw_df[column].value_counts().plot(kind='bar', color=custom_colors[i], ax=axes[i])
    axes[i].set_title(f"Bar Plot for {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Count")

# Remove any empty subplots if the number of columns is odd
if num_plots % 2 == 1:
    axes[-1].remove()

plt.tight_layout()
plt.show()
```

<h4 id="heading"><span style="background-color:#fcf3b9; color:#3d3710 ; display:fill;border-radius:5px; font-family:Georgia">📊 bar plot using seaborn</span></h4>


```python
# chose of palette for the plots from: Set1, Set2, Set3, deep, muted, pastel, bright, dark, colorblind, husl
sns.set_palette("pastel")

# Create subplots with the desired layout
num_plots = len(categorical_columns)
num_rows = num_plots // 2 + num_plots % 2
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 4 * num_rows))

# Flatten the axes array in case we have only one row of subplots
axes = axes.flatten()

# Loop through categorical columns and create bar plots using seaborn countplot
for i, column in enumerate(categorical_columns):
    sns.countplot(x=column, data=raw_df, ax=axes[i])
    axes[i].set_title(f"Bar Plot for {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Count")

# Remove any empty subplots if the number of columns is odd
if num_plots % 2 == 1:
    axes[-1].remove()

plt.tight_layout()
plt.show()
```

<h3 id="heading"><span style="background-color:#cefffb; color:#1b4946 ; display:fill;border-radius:5px; font-family:cursive"> 🛠️ ⚙️ Preprocessing </span></h3>

<h4 id="heading"><span style="background-color:#fcf3b9; color:#3d3710 ; display:fill;border-radius:5px; font-family:Georgia">🗑️ Clean the missing values </span></h4>

```python
numerical_columns = ["Income", "Children", "Cars", "Age"]

# Select only the numeric columns from the DataFrame
numeric_df = raw_df[numerical_columns]

# Fill missing values in the numeric DataFrame with the mean of each column
cleaned_raw_df = numeric_df.fillna(numeric_df.mean())
```

```python
raw_df_num_clean = raw_df
raw_df_num_clean[numerical_columns] = cleaned_raw_df
```

```python
# Check if any column has missing values
has_missing_values = raw_df_num_clean.isnull().any()

# Print only the column that having missing value
columns_with_missing_values = has_missing_values[has_missing_values]
print(columns_with_missing_values)
```
